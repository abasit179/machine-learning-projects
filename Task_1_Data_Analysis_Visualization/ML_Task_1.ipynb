#3.visualization

##Bar Plots

```python
#Bar plot by Sex
sex_counts = df['Sex'].value_counts()
plt.bar(sex_counts.index, sex_counts.values)
plt.title("Passenger Count by Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
Pclass_counts = df['Pclass'].value_counts()
plt.xticks(ticks=[1,2,3], labels=["1","2","3"])
plt.bar(Pclass_counts.index, Pclass_counts.values)
plt.title("Passenger Count by Pclass")
plt.xlabel("Pclass")
plt.ylabel("Count")
plt.show()
```

```python
Embarked_counts = df['Embarked'].value_counts()
plt.bar(Embarked_counts.index, Embarked_counts.values)
plt.title("Passenger Count by Embarked")
plt.xlabel("Embarked")
plt.ylabel("Count")
plt.show()
```

```python
Survived_counts = df['Survived'].value_counts()
plt.xticks(ticks=[0, 1], labels=["Not Survived", "Survived"])
plt.bar(Survived_counts.index, Survived_counts.values)
plt.title("Passenger Count by Survived")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.show()
```

```python
#histogram

ages = df['Age'].dropna()

plt.hist(ages, bins=20, edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
Fares = df['Fare'].dropna()

plt.hist(Fares, bins=20, edgecolor='black')
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()
```

```python
SibSps = df['SibSp'].dropna()

plt.hist(SibSps, bins=20, edgecolor='black')
plt.title("SibSp Distribution")
plt.xlabel("SibSp")
plt.ylabel("Frequency")
plt.show()
```

```python
Parchs = df['Parch'].dropna()

plt.hist(Parchs, bins=20, edgecolor='black')
plt.title("Parch Distribution")
plt.xlabel("Parch")
plt.ylabel("Frequency")
plt.show()
```

##Scatter Plot

```python
#Scatter plot used when comparing two numaric variables like Age and Fare

plt.scatter(df['Age'], df['Fare'])
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
plt.scatter(df['SibSp'], df['Parch'])
plt.title("SibSp vs Parch")
plt.xlabel("SibSp")
plt.ylabel("Parch")
plt.show()
```

##Heat Map

```python
# first of all i will select all the numerical colimns
numerical_df = df.select_dtypes(include=['int64', 'float64'])
```

```python
#computing correlation
correlation_matrix = numerical_df.corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, linecolor='gray')

plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```
